```python
import matplotlib.pyplot as plt
import numpy as np

# COS
xVector = np.arange(-5 * np.pi, 5 * np.pi, np.pi/32)
yVector = [np.cos(xElem) for xElem in xVector];
yD2Vector = [np.cos(xElem * 2) for xElem in xVector];

plt.title('R Cos(x),G Cos(2x)')
plt.axis([-6,6,-2,2])
plt.xlabel('X', color='gray')
plt.ylabel('Y', color='gray')
plt.grid(True)
plt.plot(xVector, yVector,'r', xVector, yD2Vector, 'g')
plt.show()
```
